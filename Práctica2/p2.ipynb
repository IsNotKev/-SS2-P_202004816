#### 1. Extracción y Limpieza de Datos

#### 2. Calcular el promedio de calificaciones para cada curso

Para calcular el promedio de calificaciones para cada curso se agrupa los datos por el título del curso y calcula el promedio de las calificaciones para cada grupo. Finalmente, imprime el promedio de calificaciones por curso.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('datos.csv')


# Reemplazar valores vacíos en 'Rating' por 0
df['Rating'] = df['Rating'].fillna(0)


# Convertir 'Review' y 'Duration' a float y reemplazar valores vacios por 0
df['Review'] = df['Review'].str.replace(' reviews', '').str.replace(',', '').str.extract('(\d+)').astype(float)
df['Review'] = df['Review'].fillna(0)
df['Duration'] = df['Duration'].str.extract('(\d+)').astype(float)
df['Duration'] = df['Duration'].fillna(0)

df_cleaned = df.dropna(subset=['Rating'])
average_rating = df_cleaned.groupby('Course Title')['Rating'].mean()

print("Promedio de calificaciones por curso:")
average_rating
```


Calcular el promedio de calificaciones para cada curso nos ofrece una visión general de la calidad percibida de los cursos en el conjunto de datos. Los altos promedios sugieren una satisfacción general de los estudiantes, mientras que los promedios más bajos pueden indicar áreas de mejora en la calidad del contenido o la enseñanza. Esta métrica es útil para evaluar y mejorar la oferta de cursos en plataformas de educación en línea.

#### 3. Calcular los cursos con mayor y menor rating

El código encuentra el curso con el mayor y el menor rating en el conjunto de datos, utilizando el promedio de calificaciones calculado previamente. Luego, imprime el nombre de cada curso junto con su respectivo rating más alto y más bajo.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('datos.csv')

# Reemplazar valores vacíos en 'Rating' por 0
df['Rating'] = df['Rating'].fillna(0)


# Convertir 'Review' y 'Duration' a float y reemplazar valores vacios por 0
df['Review'] = df['Review'].str.replace(' reviews', '').str.replace(',', '').str.extract('(\d+)').astype(float)
df['Review'] = df['Review'].fillna(0)
df['Duration'] = df['Duration'].str.extract('(\d+)').astype(float)
df['Duration'] = df['Duration'].fillna(0)

df_cleaned = df.dropna(subset=['Rating'])
average_rating = df_cleaned.groupby('Course Title')['Rating'].mean()

max_rating_course = average_rating.idxmax()
min_rating_course = average_rating.idxmin()

max_rating = average_rating[max_rating_course]
min_rating = average_rating[min_rating_course]

print("\nCurso con mayor rating:", max_rating_course, "con un rating de", max_rating)
print("Curso con menor rating:", min_rating_course, "con un rating de", min_rating)
```


El curso "التنسيق في مايكروسوفت إكسل | Formatting in MS Excel" recibió la calificación más alta, con un rating perfecto de 5.0, mientras que "Build social awareness content for Twitter with Canva" obtuvo la calificación más baja, con un rating de 0.0. Esto sugiere una alta valoración para el curso de MS Excel, posiblemente debido a su calidad y utilidad, mientras que el curso de Crear contenido de conciencia social para Twitter con Canva puede requerir mejoras para cumplir con las expectativas de los estudiantes.

#### 4. Calcular el porcentaje de cursos con horario flexible

El código calcula el porcentaje de cursos con horario flexible en relación con el total de cursos en el conjunto de datos y lo imprime con dos decimales.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('datos.csv')

# Reemplazar valores vacíos en 'Rating' por 0
df['Rating'] = df['Rating'].fillna(0)


# Convertir 'Review' y 'Duration' a float y reemplazar valores vacios por 0
df['Review'] = df['Review'].str.replace(' reviews', '').str.replace(',', '').str.extract('(\d+)').astype(float)
df['Review'] = df['Review'].fillna(0)
df['Duration'] = df['Duration'].str.extract('(\d+)').astype(float)
df['Duration'] = df['Duration'].fillna(0)

total_courses = len(df)
flexible_courses = len(df[df['Schedule'] == 'Flexible schedule'])
percentage_flexible_courses = (flexible_courses / total_courses) * 100

print("\nPorcentaje de cursos con horario flexible: {:.2f}%".format(percentage_flexible_courses))
```


El porcentaje de cursos con horario flexible es del 92.88%. Esto indica que la gran mayoría de los cursos en el conjunto de datos ofrecen horarios flexibles, lo que brinda a los estudiantes una mayor libertad para adaptar su aprendizaje a sus horarios y necesidades individuales. Esta alta proporción de cursos flexibles puede ser un factor atractivo para muchos estudiantes que buscan opciones de educación en línea que se ajusten a sus compromisos personales y profesionales.

#### 4. Histograma

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('datos.csv')

# Reemplazar valores vacíos en 'Rating' por 0
df['Rating'] = df['Rating'].fillna(0)


# Convertir 'Review' y 'Duration' a float y reemplazar valores vacios por 0
df['Review'] = df['Review'].str.replace(' reviews', '').str.replace(',', '').str.extract('(\d+)').astype(float)
df['Review'] = df['Review'].fillna(0)
df['Duration'] = df['Duration'].str.extract('(\d+)').astype(float)
df['Duration'] = df['Duration'].fillna(0)

# 4. Histograma de la distribución de las duraciones de los cursos
# Estilo personalizado
plt.figure(figsize=(10, 6))
n, bins, patches = plt.hist(df['Duration'], bins=10, color='skyblue', edgecolor='black', linewidth=1.2, alpha=0.7)  # Histograma con color azul cielo, bordes negros y líneas más gruesas
plt.title('Distribución de las duraciones de los cursos', fontsize=18, fontweight='bold', color='#424242')  # Color del título
plt.xlabel('Duración del curso (horas)', fontsize=14, color='#424242')  # Color de la etiqueta x
plt.ylabel('Frecuencia', fontsize=14, color='#424242')  # Color de la etiqueta y
plt.xticks(fontsize=12, color='#424242')  # Color de las marcas del eje x
plt.yticks(fontsize=12, color='#424242')  # Color de las marcas del eje y

# Fondo personalizado
plt.gca().set_facecolor('#E0E0E0')  # Color de fondo

# Agregar cantidades
for rect in patches:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%d' % int(height), ha='center', va='bottom', fontsize=10, color='black')

plt.show()
```

La mayoría de los cursos tienen una duración de 0 a 20 horas, con una disminución significativa en la frecuencia a medida que aumenta la duración del curso. Este histograma muestra la distribución de las duraciones de los cursos, con la barra más alta indicando que hay 5758 cursos con una duración de entre 0 y 20 horas. A medida que aumenta la duración del curso, las barras disminuyen drásticamente en altura, lo que sugiere una menor cantidad de cursos disponibles con largas horas de duración

#### Gráfico de cajas

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('datos.csv')

# Reemplazar valores vacíos en 'Rating' por 0
df['Rating'] = df['Rating'].fillna(0)


# Convertir 'Review' y 'Duration' a float y reemplazar valores vacios por 0
df['Review'] = df['Review'].str.replace(' reviews', '').str.replace(',', '').str.extract('(\d+)').astype(float)
df['Review'] = df['Review'].fillna(0)
df['Duration'] = df['Duration'].str.extract('(\d+)').astype(float)
df['Duration'] = df['Duration'].fillna(0)

# 5. Gráfico de cajas para visualizar la distribución de las calificaciones de los cursos por nivel de dificultad
# Estilo personalizado
plt.figure(figsize=(10, 6))
colors = ['#FF5733', '#C70039']  # Colores para cada nivel de dificultad
boxprops = dict(color='black', linewidth=1.5)  # Propiedades de las cajas
medianprops = dict(color='black', linewidth=2)  # Propiedades de la línea de la mediana

df.boxplot(column='Rating', by='Level', grid=False, ax=plt.gca(), boxprops=boxprops, medianprops=medianprops, patch_artist=True, notch=True, flierprops=dict(marker='o', markerfacecolor='red', markersize=8))  # Gráfico de cajas
plt.title('Distribución de las calificaciones por nivel de dificultad', fontsize=18, fontweight='bold', color='#424242')  # Título
plt.suptitle('')  # Eliminar título predeterminado
plt.xlabel('Nivel de dificultad', fontsize=14, color='#424242')  # Etiqueta del eje x
plt.ylabel('Calificación', fontsize=14, color='#424242')  # Etiqueta del eje y
plt.xticks(fontsize=12, color='#424242')  # Tamaño y color de las marcas del eje x
plt.yticks(fontsize=12, color='#424242')  # Tamaño y color de las marcas del eje y

# Aplicar colores personalizados a las cajas
for patch, color in zip(plt.gca().artists, colors):
    patch.set_facecolor(color)

# Fondo personalizado
plt.gca().set_facecolor('#E0E0E0')  # Color de fondo

plt.show()
```

El gráfico de cajas que muestra la distribución de las calificaciones de los cursos por nivel de dificultad revela lo siguiente: **Nivel Avanzado**: Las calificaciones son similares a las del nivel intermedio, con una mediana alrededor de 4.0. Sin embargo, hay algunos valores atípicos bajos en este nivel. **Nivel Intermedio:** Las calificaciones también tienen una mediana cercana a 4.0. La mayoría de las calificaciones se encuentran dentro del rango intercuartílico (IQR). **Nivel Principiante:** Las calificaciones tienen una mediana ligeramente inferior, alrededor de 3.75. Al igual que en los otros niveles, hay algunas calificaciones atípicas. En general, los cursos en todos los niveles tienen calificaciones positivas, pero el nivel avanzado puede tener una mayor variabilidad debido a los valores atípicos bajos.
